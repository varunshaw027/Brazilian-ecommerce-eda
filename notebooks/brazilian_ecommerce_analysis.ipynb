```python
Q1 = payments['payment_value'].quantile(0.25)

Q3 = payments['payment_value'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR

upper_bound = Q3 + 1.5 * IQR

outliers = payments[
    (payments['payment_value'] < lower_bound) |
    (payments['payment_value'] > upper_bound)
]

print("Upper Bound:", upper_bound)

print("Total Outliers:", outliers.shape[0])

outliers[['payment_value']].describe()
```

```python
total_transactions = payments.shape[0]

outlier_transactions = outliers.shape[0]

outlier_percentage = (
    outlier_transactions / total_transactions
) * 100

print("Total Transactions:", total_transactions)

print("Outlier Transactions:", outlier_transactions)

print("Outlier Percentage:", round(outlier_percentage, 2), "%")
```

```python
normal_payments = payments[
    payments['payment_value'] <= upper_bound
]

print("Normal Payments")

print(
    normal_payments['payment_value'].describe()
)

print("\nOutlier Payments")

print(
    outliers['payment_value'].describe()
)
```

```python
orders = pd.read_csv(
    '../data/orders.csv',
    header=None
)

customers = pd.read_csv(
    '../data/customers.csv',
    header=None
)


orders.columns = [
    'order_id',
    'customer_id',
    'order_status',
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
```

```python
customers.columns = [
    'customer_id',
    'customer_unique_id',
    'customer_zip_code_prefix',
    'customer_city',
    'customer_state'
]
```

```python
print(orders.columns)

print(customers.columns)
```

```python
ecommerce = orders.merge(
    payments,
    on='order_id',
    how='inner'
)

ecommerce = ecommerce.merge(
    customers,
    on='customer_id',
    how='inner'
)

print(ecommerce.shape)

ecommerce.head()
```

```python
ecommerce.info()
```

```python
state_revenue = ecommerce.groupby(
    'customer_state'
)['payment_value'].sum().sort_values(
    ascending=False
)

print(state_revenue.head(10))
```

```python
plt.figure(figsize=(12,6))

state_revenue.head(10).plot(
    kind='bar'
)

plt.title("Top 10 States by Revenue")

plt.xlabel("State")

plt.ylabel("Revenue")

plt.show()
```

```python
payment_method = ecommerce.groupby(
    'payment_type'
)['payment_value'].sum().sort_values(
    ascending=False
)

print(payment_method)
```

```python
plt.figure(figsize=(8,5))

payment_method.plot(
    kind='bar'
)

plt.title("Revenue by Payment Type")

plt.xlabel("Payment Type")

plt.ylabel("Revenue")

plt.show()
```

```python
ecommerce['order_purchase_timestamp'] = pd.to_datetime(
    ecommerce['order_purchase_timestamp']
)

ecommerce['month'] = ecommerce[
    'order_purchase_timestamp'
].dt.to_period('M')

monthly_revenue = ecommerce.groupby(
    'month'
)['payment_value'].sum()

print(monthly_revenue)
```

```python
plt.figure(figsize=(14,6))

monthly_revenue.plot()

plt.title("Monthly Revenue Trend")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.show()
```

```python
city_revenue = ecommerce.groupby(
    'customer_city'
)['payment_value'].sum().sort_values(
    ascending=False
)

print(city_revenue.head(10))
```

```python
plt.figure(figsize=(12,6))

city_revenue.head(10).plot(
    kind='bar'
)

plt.title("Top 10 Cities by Revenue")

plt.xlabel("City")

plt.ylabel("Revenue")

plt.show()
```

```python
payment_avg = ecommerce.groupby(
    'payment_type'
)['payment_value'].mean().sort_values(
    ascending=False
)

print(payment_avg)
```

```python
ecommerce['order_delivered_customer_date'] = pd.to_datetime(
    ecommerce['order_delivered_customer_date']
)

ecommerce['order_estimated_delivery_date'] = pd.to_datetime(
    ecommerce['order_estimated_delivery_date']
)
```

```python
ecommerce['delivery_delay'] = (
    ecommerce['order_delivered_customer_date']
    -
    ecommerce['order_estimated_delivery_date']
).dt.days
```

```python
print(
    ecommerce['delivery_delay'].describe()
)
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    ecommerce['delivery_delay'],
    bins=50
)

plt.title("Delivery Delay Distribution")

plt.xlabel("Delay Days")

plt.ylabel("Frequency")

plt.show()
```

```python
state_delay = ecommerce.groupby(
    'customer_state'
)['delivery_delay'].mean().sort_values(
    ascending=False
)

print(state_delay.head(10))
```

```python
total_revenue = ecommerce['payment_value'].sum()

average_order = ecommerce['payment_value'].mean()

total_orders = ecommerce['order_id'].nunique()

total_customers = ecommerce['customer_id'].nunique()

print("Total Revenue:", total_revenue)

print("Average Order Value:", average_order)

print("Total Orders:", total_orders)

print("Total Customers:", total_customers)
```

# Revenue Insights

## Key Findings
- São Paulo (SP) generated the highest ecommerce revenue.
- Revenue concentration is heavily skewed toward a few states.
- Credit cards dominate ecommerce transactions.
- Average order value indicates mid-range consumer spending behavior.

## Business Recommendations
- Expand logistics infrastructure in SP.
- Target high-performing regions with premium campaigns.
- Improve customer acquisition in low-performing states.

# Delivery Performance Insights

## Key Findings
- Certain states show higher delivery delays.
- Delivery performance varies significantly by geography.
- Late deliveries may impact customer satisfaction and retention.

## Business Recommendations
- Optimize logistics routes in delayed regions.
- Evaluate warehouse distribution strategy.
- Improve delivery partner performance tracking.

# Payment Behavior Insights

## Key Findings
- Credit cards are the most commonly used payment method.
- Installment-based payments contribute significantly to revenue.
- Payment behavior may indicate customer affordability preferences.

## Business Recommendations
- Promote installment payment offers.
- Introduce loyalty incentives for repeat customers.
- Optimize checkout experience for dominant payment methods.
